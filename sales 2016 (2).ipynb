```python
grouped = df.groupby(['Brand']).aggregate(
    {'SalesQuantity': ['sum']})
print(grouped.head())
```

```python
dfofbrandsales.columns = ['SalesQuantity']
dfofbrandsales = dfofbrandsales.reset_index()
```

```python
top_30_sale_brand=dfofbrandsales.sort_values(by='SalesQuantity',ascending=False,ignore_index=True)
```

```python
ba=top_30_sale_brand[1:31]
ba.head(5)
```

```python
g=sns.barplot(x =ba['Brand'],y = ba['SalesQuantity'],data = ba,)
g.set_xticklabels(ba['Brand'],rotation=90)
([])
plt.title("Top 30 Brand Sales Quantity")
```

```python
#these are the some most pouplar brand 
```
